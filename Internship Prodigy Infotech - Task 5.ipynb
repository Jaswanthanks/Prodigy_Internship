```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:\\UK_Accident.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(inplace = True) 
```

```python
df = pd.DataFrame(df)
```

```python
print(df.columns)
```

```python
df['Accident_Severity'] = df['Accident_Severity'].astype('category')
df['Weather_Conditions'] = df['Weather_Conditions'].astype('category')
df['Road_Surface_Conditions'] = df['Road_Surface_Conditions'].astype('category')
df['Light_Conditions'] = df['Light_Conditions'].astype('category')
df['Year'] = df['Year'].astype(int)
```

# Effect by Weather Condition 

```python
weather_condition = df['Weather_Conditions'].value_counts()

plt.figure(figsize = (10,6))
plt.barh(weather_condition.index , weather_condition.values , color = "skyblue" , edgecolor = 'black')
plt.title("Distribution of accident by weather condition")
plt.xlabel('counts')
plt.ylabel("weather_condition")
plt.gca().invert_yaxis()
plt.show()
```

# Effects by Road Surface Condtion

```python
road_sur_condition = df['Road_Surface_Conditions'].value_counts()

plt.figure(figsize = (10,6))
plt.barh(road_sur_condition.index ,road_sur_condition.values, color = "tomato" , edgecolor = "black")
plt.title("Accident affected by the road surface condition")
plt.xlabel("counts")
plt.ylabel("Road_Surface_Condition")
plt.gca().invert_yaxis()
plt.show()
```

# Effects By Light Condtion

```python
light_condition = df['Light_Conditions'].value_counts()

plt.figure(figsize = (10,6))
plt.barh(light_condition.index ,light_condition.values, color = "orchid" , edgecolor = "black")
plt.title("Accident affected by the light condition")
plt.xlabel("counts")
plt.ylabel("light condition")
plt.gca().invert_yaxis()
plt.show()
```
