```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("E:\\dataset\\titanic.csv")
df.head()
```

```python
df['sex'].value_counts().plot(kind='bar', color=['Orange', 'red'])
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.hist(df['age'], bins=5, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
